```python
movies_df = pd.read_pickle("../main/data/movies_df.pickle")
```

```python
movies_df.shape
```

```python
movies_df[movies_df.title == 'Avatar']
```

```python
movies_df.rename(columns={"id": "tmdb_id"}, inplace=True)
```

```python
links.rename(columns={"tmdbId": "tmdb_id"}, inplace=True)
```

```python
df = movies_df.merge(links, on='tmdb_id')
```

```python
df.head()
```

```python
# drop duplicates and missing movieIds
df = df.drop_duplicates(subset="tmdb_id")
df = df.dropna(subset="movieId")

df.head()
```

```python
df.shape
```

```python
df_full = ratings.merge(df, on='movieId')
```
